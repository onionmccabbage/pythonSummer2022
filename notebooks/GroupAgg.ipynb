# Grouping and Aggregating Data

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# import some data
df = pd.read_csv('../Data/Data For Analysis_Salaries.csv')
df.tail()
```

```python
df.describe()
df['salary'].dtype
df.columns
# df.cumsum()
df['service']
# grouping data
df_rank = df.groupby(['rank']) # NB we provide a LIST and get a groupby object
df_rank.max()
# how can we find the mean salary for each proessor
# df_mean_prof = df.groupby(['rank'])[['salary', 'service']].mean()
cols = ['salary', 'service']
df_mean_prof = df_rank[cols].mean()
df_mean_prof
```

```python
# sincee we have a df indexed by number, lets compare loc and iloc
df.loc[10:20] # from 10 to 20 inclusive
df.iloc[10:20].head() # from 10, stop before 20
```

```python
df.sort_values( by='service', inplace=True )
df
```

```python
df.groupby('discipline')[['rank']].count()
```

```python
# use the agg() method to aggregate the min, mean etc.
cols = ['salary', 'service', 'rank']
factors = ['min', 'max', 'mean', 'sum']
df[cols].agg(factors)
```

```python
# unique members
df.agg({'rank':['nunique']}) # number of unique members
```
